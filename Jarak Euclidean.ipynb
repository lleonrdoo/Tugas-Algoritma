```python
import math
import matplotlib.pyplot as plt

# Fungsi untuk menghitung jarak Euclidean
def euclidean_distance(p1, p2):
    return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)

# Memasukkan titik awal
x0, y0 = map(float, input("Masukkan koordinat titik awal (x0 y0): ").split())

# Memasukkan 3 titik lokasi lainnya
points = []
for i in range(2):  # Perbaikan dari 2 ke 3 untuk menerima 3 titik
    x, y = map(float, input(f"Masukkan koordinat titik ke-{i+1} (x y): ").split())
    points.append((x, y))

# Menentukan titik terdekat (titik kedua)
distances = [euclidean_distance((x0, y0), point) for point in points]
min_distance = min(distances)
second_point = points[distances.index(min_distance)]

# Menentukan titik berikutnya yang lebih jauh dari titik awal melalui titik kedua (titik ketiga)
remaining_points = [point for point in points if point != second_point]
distances_via_second = [euclidean_distance(second_point, point) for point in remaining_points]
third_point = remaining_points[distances_via_second.index(max(distances_via_second))]

# Menghitung total jarak
total_distance = (euclidean_distance((x0, y0), second_point) +
                  euclidean_distance(second_point, third_point) +
                  euclidean_distance(third_point, (x0, y0)))

# Output hasil
print(f"Titik awal: A({x0}, {y0})")
print(f"Titik kedua: B{second_point}")
print(f"Titik ketiga: C{third_point}")
print(f"Jumlah jarak total: {total_distance}")
print(f"Rute: A -> B -> C -> A")

# Plotting titik-titik dan rute
plt.figure(figsize=(8, 8))
plt.scatter(x0, y0, color='red', label='Titik Awal (A)')
plt.scatter(*zip(*points), color='blue', label='Titik Lokasi')
plt.scatter(*second_point, color='green', label='Titik Kedua (B)', marker='x')
plt.scatter(*third_point, color='purple', label='Titik Ketiga (C)', marker='x')

# Plotting garis rute
plt.plot([x0, second_point[0]], [y0, second_point[1]], color='black', linestyle='--')
plt.plot([second_point[0], third_point[0]], [second_point[1], third_point[1]], color='black', linestyle='--')
plt.plot([third_point[0], x0], [third_point[1], y0], color='black', linestyle='--')

# Labeling titik-titik
plt.text(x0, y0, ' A', fontsize=12, ha='right')
plt.text(second_point[0], second_point[1], ' B', fontsize=12, ha='right')
plt.text(third_point[0], third_point[1], ' C', fontsize=12, ha='right')

plt.xlabel('Koordinat X')
plt.ylabel('Koordinat Y')
plt.title('Rute Berdasarkan Jarak Euclidean')
plt.legend()
plt.grid(True)
plt.show()
```
