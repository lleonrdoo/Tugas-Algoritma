```python
import math
import matplotlib.pyplot as plt

#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#$$$$$$$$$$$$$$$$$$$$$$$$$$$  PROSES PEMBENTUKAN FUNGSI  $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#$$$$$$$$$$$$$$$$$$$$$$$  JARAK EUCLID DAN NEAREST NEIGHBOR  $$$$$$$$$$$$$$$$$$$$$$$$$$$$

# Fungsi untuk menghitung jarak Euclidean
def jarak_euclidean(p1, p2):
    return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2) # p1 dan p2 adalah data bertipe array

# Fungsi untuk menemukan titik terdekat
def find_nearest_neighbor(current_point, points):
    titik_terdekat = None # nearest_point = titik_terdekat
    jarak_minimal = float('inf')
    for point in points:
        jarak = jarak_euclidean(current_point, point)
        if jarak < jarak_minimal:
            jarak_minimal = jarak
            titik_terdekat = point
    return titik_terdekat, jarakminimal



#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$  PROSES INPUT DAN OUTPUT  $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$

# Pengguna memasukkan titik awal dan titik lainnya
x0, y0 = map(float, input("Masukkan koordinat titik awal (x0 y0): ").split())
titik_awal = (x0, y0)

# Memasukkan jumlah titik lainnya
n = int(input("Masukkan jumlah titik lainnya: "))
titik_lain = []
for i in range(n):
    x, y = map(float, input(f"Masukkan koordinat titik ke-{i+1} (x y): ").split())
    titik_lain.append((x, y))

# Algoritma Nearest Neighbor untuk TSP
titik_semua = [titik_awal] + titik_lain
rute = [titik_awal]
current_point = titik_awal
total_jarak = 0

while titik_lain:
    nearest_point, distance = find_nearest_neighbor(current_point, titik_lain)
    rute.append(nearest_point)
    total_jarak += distance
    current_point = nearest_point
    titik_lain.remove(nearest_point)

# Tambahkan jarak kembali ke titik awal
total_jarak = total_jarak + jarak_euclidean(current_point, titik_awal)
rute.append(titik_awal)

# Output hasil
print(f"Rute terbaik: {rute}")
print(f"Jumlah jarak total: {total_jarak}")



#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$  PROSES PLOTTING RUTE  $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$

# Plot rute menggunakan matplotlib
x_values = [p[0] for p in rute]
y_values = [p[1] for p in rute]

plt.plot(x_values, y_values, marker='o')
for i, titik in enumerate(rute):
    plt.text(titik[0], titik[1], f'{i}', fontsize=12, ha='right')

plt.title('Rute Terbaik dengan Nearest Neighbor TSP')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
