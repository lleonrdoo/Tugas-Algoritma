#### Soal : f(x) = x - sin x

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Pendefinisian fungsi 
def f(x):
    return x * np.sin(x) - 1

# np.zeros digunakan untuk membuat array yang memiliki nilai '0'
# '25' berarti array memiliki index sampai 25
a = np.zeros(25)
b = np.zeros(25)
c = np.zeros(25)

# Ini merupakan nilai awal dari variabel a dan b 
a[0] = 0
b[0] = 2

# Ini merupakan nilai awal dari variabel c
c[0] = (a[0] + b[0]) / 2

results = []

tolerance = 1e-6  # Toleransi yang ditentukan 0.000001
max_iter = 1000   # Jumlah iterasi maksimum

for n in range(max_iter):
    # - b[n] adalah nilai titik akhir interval pada iterasi ke-n.
    # - a[n] adalah nilai titik awal interval pada iterasi ke-n.
    # - b[n] - a[n] adalah panjang interval pada iterasi ke-n.
    # - abs(b[n] - a[n]) mengambil nilai absolut dari panjang interval tersebut, sehingga pasti positif.
    # - tolerance adalah toleransi yang ditetapkan untuk menentukan konvergensi. 
    #   Jika perbedaan antara b dan a lebih kecil dari toleransi, maka metode dianggap telah konvergen.
    if abs(b[n] - a[n]) < tolerance:
        print(f"Konvergensi tercapai setelah {n} iterasi.")
        break
    
    # Sesuai algoritma iterasi metode biseksi:
    # - Jika f(a).f(c) < 0, maka a[1] = a dan b[1] = c
    # - Jika f(a).f(c) > 0, maka a[1] = c dan b[1] = b
    if f(a[n]) * f(c[n]) < 0:
        a[n+1] = a[n]
        b[n+1] = c[n]
    elif f(a[n]) * f(c[n]) > 0:
        a[n+1] = c[n]
        b[n+1] = b[n]
    elif f(a[n]) * f(c[n]) == 0:
        print(f"Maka akar persamaan adalah c = {c[n]} ")
    else:
        break
        
    c[n+1] = (a[n+1] + b[n+1]) / 2
    results.append([a[n], b[n], c[n], f(c[n]), f(a[n])])

# Syntax untuk menampilkan tabel Nilai a, Nilai b, Nilai c, dan f(c)
df = pd.DataFrame(results, columns=['Nilai a', 'Nilai b', 'Nilai c', 'f(c)', 'f(a)'])

print(df)

# Print akar persamaan
if abs(b[n] - a[n]) < tolerance:
    print(f"Diketahui bahwa nilai terakhir dari variabel a = {a[n]} dan variabel b = {b[n]}")
    print(f"Hampiran akar persamaan f(x) =  x.sin(x) - 1 ditemukan pada nilai c = {c[n]}")
else:
    print("Metode biseksi tidak konvergen setelah jumlah iterasi maksimum.")

# Plotting fungsi
# Membuat array x_values yang berisi 100 titik antara 0 dan 2 secara merata. Ini digunakan sebagai nilai x pada plot.
x_values = np.linspace(0, 2, 100)
# Menghitung nilai fungsi f(x) untuk setiap nilai x dalam x_values.
y_values = f(x_values)

# Membuat plot garis dengan nilai x_values sebagai sumbu x dan y_values sebagai sumbu y, dengan label 'f(x)'.
plt.plot(x_values, y_values, label='f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot Fungsi f(x)')
plt.grid(True)

# Menandai titik c pada grafik
# Membuat titik scatter pada koordinat (c[n], f(c[n])) dalam warna merah dengan label 'Titik c'.
plt.scatter(c[n], f(c[n]), color='red', label='Titik c')

# Menandai titik awal a[0] dan b[0]
plt.scatter(a[0], f(a[0]), color='blue', label='Titik a[0]')
plt.scatter(b[0], f(b[0]), color='green', label='Titik b[0]')

plt.legend()
plt.show()
```
