```python
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
# Latihan 5.21.2
# Penggunaan Properti pada Pandas
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

# diperlukan inisialisasi library pandas kembali
import pandas as pd
data31 = pd.read_csv('C:/Users/santi/bunga_iris.csv')

# berikut pemanggilan variabel dengan urutan baris 1-4
print("Pernyataan (1) : ")
print(data31.head(4))
print('\n\n')

# berikut pemanggilan variabel berdasarkan nama-nama kolomnya
print("Pernyataan (2) :")
print(data31.columns)
print('\n\n')

# berikut pemanggilan variabel berdasarkan inisial string kolom datanya
print("Pernyataan (3) :")
print(data31['sepal length (cm)'])
print('\n\n')

# berikut pemanggilan variabel berdasarkan inisial string kolom dan urutan index barisnya
print("Pernyataan (4) :")
print(data31['sepal length (cm)'][0])
print('\n\n')

# berikut pemanggilan variabel berdasarkan inisial string kolom dan urutan index barisnya
print("Pernyataan (5) :")
print(data31['sepal length (cm)'][1:5])
print('\n\n')

# berikut pemanggilan variabel berdasarkan inisial string dua buah kolom dan urutan index baris kedua kolomnya
print("Pernyataan (6) :")
print(data31[['sepal length (cm)','sepal width (cm)']][1:5])
print('\n\n')

# berikut pemanggilan variabel berdasarkan urutan baris 
# dan kolom-kolom pada baris tersebut akan diurutkan secara vertikal
# catatan : hal ini berlaku jika hanya terdapat 1 index yang diminta
print("Pernyataan (7) :")
print(data31.iloc[1])
print('\n\n')

# berikut pemanggilan variabel berdasarkan urutan baris sesuai index baris yang diminta
print("Pernyataan (8) :")
print(data31.iloc[1:5])
print('\n\n')

# berikut pemanggilan variabel berdasarkan inisial string tertentu yang dioperasikan operator perbandingan pada data tertentu
print("Pernyataan (9) :")
print(data31.loc[data31['sepal width (cm)']==3.0])
```

```python
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
# Latihan 5.21.6
# Penyaringan Data Berdasarkan Syarat dan Regex
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

import pandas as pd
import re
data31 = pd.read_csv('C:/Users/santi/bunga_iris.csv')

# berikut penyaringan data berdasarkan inisial string yang dituju dengan fungsi 'loc'dan dengan fungsi 'str.contains()'
# fungsi 'regex = True' digunakan untuk mengurutan karakter yang membentuk pola untuk menemukan, mengganti, dan mengekstrak data tekstual 
# namun, karena didalam inisial string yang dituju hanya terdapat data integer, makan 'str.contains()'' mengalami error
print(data31.loc[data31['sepal length (cm)'].str.contains(3.0|3.5 , regex=True)])
```

```python
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
# Latihan 5.21.6
# Penyaringan Data Berdasarkan Syarat, Regex, dan Flags
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

import pandas as pd
import re
data31 = pd.read_csv('C:/Users/santi/bunga_iris.csv')

# berikut penyaringan data berdasarkan inisial string yang dituju dengan fungsi 'loc'dan dengan fungsi 'str.contains()'
# fungsi 'regex = True' digunakan untuk mengurutan karakter yang membentuk pola untuk menemukan, mengganti, dan mengekstrak data tekstual 
# fungsi 'flags=re.I' digunakan sebagai argumen ke metode pencarian untuk mengaktifkan pencarian yang tidak sensitif huruf
# namun, karena didalam inisial string yang dituju hanya terdapat data integer, makan 'str.contains()'' mengalami error
print(data31.loc[data31['sepal length (cm)'].str.contains(3.0|3.5 , flags=re.I , regex=True)])
```

## 5.22.2 Properti Pembuatan Grafik dengan Matplotlib 

```python
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
# Latihan 5.22.2
# Membentuk Grafik dengan Properti pada Library Matoplotlib (1)
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data31 = pd.read_csv('C:/Users/santi/bunga_iris.csv')

# berikut inisialisasi variabel dengan data variabel sesuai inisialisasi kolom string yang dituju
x = data31['sepal length (cm)']
y = data31['petal length (cm)']
x1 = data31['sepal width (cm)']
y1 = data31['petal width (cm)']

# berikut pembentukan sumbu koordinat yang berukuran 15 x 15 inch dengan resolusi gambar 500 dpi
# fungsi 'figure()' digunakan untuk menggambarkan bentuk dari koordinatnya
# fungsi 'figsize()' digunakan untuk menentukan ukuran gambar
# fungsi 'dip' digunakan untuk menentukan resolusi gambar
plt.figure(figsize=(15,15), dpi=500) 

# berikut pembentukan linier pertama dengan nama, warna, penanda, dan bentuk garis tertentu
# fungsi 'label' digunakan untuk menamai linier yang diminta
# fungsi 'color' digunakan untuk menentukan warna linier
# fungsi 'linewidth' digunakan untuk ketebalan linier
# fungsi 'marker' digunakan untuk menentukan penanda pada titik yang dituju
# fungsi 'makersize' digunakan unutuk menentukan ukuran penanda
# fungsi 'markeredgecolor' digunakan untuuk menentukan warna penanda
# fungsi 'markersize' digunakan untuk menentukan ukuran penanda
plt.plot(x,y, label = 'Grafik length' ,  color = 'blue' , 
         linewidth = 3 , marker = '*' , markersize = 15 , 
         markeredgecolor = 'orange' , linestyle='-')

# berikut pembentukan garis liniear kedua dengan nama, warna, penanda, dan bentuk garis tertentu
# fungsi 'label' digunakan untuk menamai linier yang diminta
# fungsi 'linewidth' digunakan untuk ketebalan linier
# fungsi 'marker' digunakan untuk menentukan penanda pada titik yang dituju
# fungsi 'makersize' digunakan unutuk menentukan ukuran penanda
# fungsi 'markeredgecolor' digunakan untuk menentukan warna penanda
plt.plot(x1,y1, label = 'Garik width', marker = '*' , 
         markersize = 15 , markeredgecolor = 'orange' )

# berikut penamaan koordinat dengan fungsi 'title()'
# fungsi 'title()' digunakan untuk memasukan data string dan ukuran font
# fungsi 'fontdict' digunakan untuk menentukan ukuran font
plt.title('Grafik Bunga Iris ', fontdict = {'fontname' : 'DejaVu Sans' , 'fontsize' : 12})

# berikut inisialisasi sumbu x dan y dengan penamaan menggunakan data string
# fungsi 'fontdict' digunakan untuk menentukan ukuran font
plt.xlabel('X', fontdict={'fontname' : 'DejaVu Sans' , 'fontsize' : 6})
plt.ylabel('Y', fontdict={'fontname' : 'DejaVu Sans' , 'fontsize' : 6})

# berikut pembentukan output dari label x dan y
plt.xticks(data31['sepal length (cm)'])
plt.yticks(data31['petal length (cm)'])
plt.legend()

# berikut pembentukan output dari label x1 dan y1
plt.xticks(data31['sepal width (cm)'])
plt.yticks(data31['petal width (cm)'])
plt.legend()

# berikut penyimpanan figure pada 'grafik.png' dengan resolusi 500 dpi
plt.savefig('grafik.png' , dpi=500)

# berikut fungsi 'show()' yang digunakan untuk mengeluarkan output figure
plt.show()
```

```python
# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
# Latihan 5.22.3
# Membentuk Histogram Sederhana dengan Library Matoplotlib (1)
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data31 = pd.read_csv('C:/Users/santi/bunga_iris.csv')

# berikut inisialisasi variabel dengan menginput data yang dituju sesuai dengan kolom yang diminta
sepal = data31['sepal length (cm)']
petal = data31['petal length (cm)']

# berikut inisialisasi variabel dengan menginput data yang dituju sesuai dengan kolom yang diminta
# namun terdapat operator perbandingan sebagai pengelompokan data kolom yang diminta
sepal1 = data31['sepal width (cm)']
petal1 = data31['petal width (cm)']

# berikut inisialisasi sumbu x dan y dengan penamaan menggunakan data string
# fungsi 'fontdict' digunakan untuk menentukan ukuran font
plt.xlabel('Sepal (cm)')
plt.ylabel('Petal (cm)')

# berikut penggunaan fungsi 'bar()' untuk membentuk histogram
print(plt.bar(sepal,sepal1))
print(plt.bar(petal,petal1))
```

## 5.22.4 Properti Pembuatan Histogram dengan Matplotlib 

```python
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
# Latihan 5.22.4
# Membentuk Histogram dengan Properti pada Library Matoplotlib
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data31 = pd.read_csv('C:/Users/santi/bunga_iris.csv')

# berikut inisialisasi variabel dengan menginput data integer
x = [-5,-3,-1,0,1,3,5,6]
y = [31,13,3,1,1,7,21,31]

# berikut pembentukan sumbu koordinat yang berukuran 12 x 12 inch dengan resolusi gambar 500 dpi
# fungsi 'figure()' digunakan untuk menggambarkan bentuk dari koordinatnya
# fungsi 'figsize()' digunakan untuk menentukan ukuran gambar
plt.figure(figsize=(5,5))

# berikut inisialisasi variabel dengan fungsi 'bar()' beserta menentukan warnanya
hist = plt.bar(x,y,color='blue')
pola = ['/','/','/','/','x','x','x','x']

# berikut digunakan fungsi 'for' untuk mengurutkan variabel hist
for bar in hist :
    # fungsi 'set_hatch()' digunakan untuk pola di dalam batang
    bar.set_hatch(pola.pop(0))

# berikut fungsi 'show()' yang digunakan untuk mengeluarkan output figure
plt.show()
```
